#### Problem 1

Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.


```python
class Line:

    def __init__(self, coor1, coor2):
        self.coor1 = coor1
        self.coor2 = coor2

    def distance(self):
        dist = (
            (self.coor1[0] - self.coor2[0]) ** 2 + (self.coor1[1] - self.coor2[1]) ** 2
        ) ** 0.5
        return dist

    def slope(self):
        if self.coor1[0] - self.coor2[0] == 0:
            return None

        slpe = (self.coor1[1] - self.coor2[1]) / (self.coor1[0] - self.coor2[0])
        return slpe
```

```python
# EXAMPLE OUTPUT

coordinate1 = (3, 2)  # tuple
coordinate2 = (8, 10)  # tuple

li = Line(coordinate1, coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

#### Problem 2


Fill in the class


```python
class Cylinder:

    def __init__(self, height=1, radius=1):
        self.height = height
        self.radius = radius

    def volume(self):
        return self.radius**2 * 3.14 * self.height

    def surface_area(self):
        return (self.radius**2 * 3.14) * 2 + (self.radius * 2 * 3.14 * self.height)
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2, 3)
```

```python
c.volume()
```

```python
c.surface_area()
```
