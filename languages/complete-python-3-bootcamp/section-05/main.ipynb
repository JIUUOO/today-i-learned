```python
mylist = range(0, 10)

for num in mylist:
    # Check for even
    if num % 2 == 0:
        print(num)
    else:
        print(f"Odd Number: {num}")
```

```python
d = {"k1": 1, "k2": 2, "k3": 3}

for item in d.values():
    print(item)
# 1
# 2
# 3
```

```python
d = {"k1": 1, "k2": 2, "k3": 3}

for item in d.items():
    print(item)
# ('k1', 1)
# ('k2', 2)
# ('k3', 3)
```
