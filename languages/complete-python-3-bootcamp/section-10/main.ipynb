```python
try:
    f = open("testfile", "w")
    f.write("Write a test line")
except TypeError:
    print("There was a type error!")
except:
    print("All other exceotions!")
finally:
    print("I always run")
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input("Please provide number: "))
        except:
            print("Whoops! That is not a number")
            continue
        else:
            print("Yes thank you")
            break
        finally:
            print("End of try/except/finally")
            print("I will always run at the end!")


ask_for_int()
```
