```python
def myfunc(**kwargs):
    print(kwargs)

    if "fruit" in kwargs:
        print("My fruit of choice is {}".format(kwargs["fruit"]))

    else:
        print("I did not find any fruit here")


myfunc(fruit="apple", veggie="lettuce")
"""
{'fruit': 'apple', 'veggie': 'lettuce'}
My fruit of choice is apple
"""
```

```python
def is_prime(x):
    if x in (0, 1):
        return False
    for i in range(2, x):
        if x % i == 0:
            return False

    return True


def count_primes(x):
    count = 0
    for i in range(2, x + 1):
        if is_prime(i):
            count += 1
    return count
```
