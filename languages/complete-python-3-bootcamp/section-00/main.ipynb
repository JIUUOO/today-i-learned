```python
my_list = [1, 2, 3]
my_list = ["STRING", 100, 23.2]
len(my_list)  # 3

mylist = ["one", "two", "three"]
mylist[0]
mylist[1:]

another_list = ["four", "five"]
new_list = (
    mylist + another_list
)  # ['one', 'two', 'three', 'four', 'five'] (contatenation)
new_list[0] = "ONE ALL CAPS"
new_list  # ['ONE ALL CAPS', 'two', 'three', 'four', 'five']

new_list.append("six")  # add
new_list  # ['ONE ALL CAPS', 'two', 'three', 'four', 'five', 'six']

new_list.pop()  # 'six' (in-place function)
new_list  # ['ONE ALL CAPS', 'two', 'three', 'four', 'five']
```

```python
new_list = ["a", "e", "x", "b", "c"]
new_list.sort()  # in-place method
new_list  # ['a', 'b', 'c', 'e', 'x']
```

```python
d = {"k1": 100, "k2": 200}
d  # {'k1': 100, 'k2': 200}
d["k3"] = 300
d  # {'k1': 100, 'k2': 200, 'k3': 300}

d["k1"] = "NEW VALUE"
d  # {'k1': 'NEW VALUE', 'k2': 200, 'k3': 300}
```

```python
d = {"k1": 100, "k2": 200, "k3": 300}
d.keys()  # dict_keys(['k1', 'k2', 'k3'])
d.values()  # dict_values([100, 200, 300])
d.items()  # dict_items([('k1', 100), ('k2', 200), ('k3', 300)])
```
