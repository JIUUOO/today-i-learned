```python
import itertools

# 4! permutations of 1,2,3,4, as a list
list(itertools.permutations([1, 2, 3, 4]))
```

```python
# 3!/2! unique permutations of 1,1,2 as a set
set(itertools.permutations([1, 1, 2]))
```

```python
# 3!/2! unique permutations of 1,1,2 as a list
list(itertools.permutations([1, 1, 2]))  # <- overcounted
```

```python
# 5C3 choosing three numbers from [1,2,3,4,5]
list(itertools.combinations([1, 2, 3, 4, 5], 3))
```
